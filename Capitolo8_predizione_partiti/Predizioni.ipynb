```python
import pandas as pd
import os
import csv
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import time
from sklearn.metrics import recall_score
```

### Predictions

```python
G = nx.read_edgelist("./EDGES_FILE.csv", create_using=nx.DiGraph, nodetype=int)
H = nx.read_edgelist("./EDGES_FILE.csv", create_using=nx.DiGraph, nodetype=int)
```

```python
for (x, y) in G.edges():
    if x == y:
        if H.has_node(x):
            H.remove_node(x)
G = H
```

### Using first method

```python
total_preds = []
all_accuracies = []
all_preds = []
all_recalls = []
iterations = []
counts = []
for i in range(len(os.listdir("./id_numerici/")) - 1):
    
    # Get the two months to compare's folders
    folder = sorted(os.listdir("./id_numerici/"))[i]
    next_folder = sorted(os.listdir("./id_numerici/"))[i + 1]
    
    s_time = time.time()

    # Reading VLs from month's folder
    obj = pd.read_pickle(f"./id_numerici/{folder}/results/log_trial_0_LPStates.pickled")
    
    ### Reading initial parties of nodes exluding nodes without links
    comms = []
    for node in sorted(list(G.nodes)):
        if(1 in obj[0][1][node - 1]):
            comms.append(obj[0][1][node - 1].index(1)) # + 1 forse
        else: comms.append(-1)
    data = comms
    
    ### Creating community VLs
    VLs = []
    for res in obj:
        VL = []
        for comm in sorted(pd.Series(data).unique()):
            if(comm != -1):
                node_list = []
                for i in sorted(list(G.nodes())):
                    if data[sorted(list(G.nodes())).index(i)] == comm:
                        node_list.append(res[1][i - 1])
                series = [pd.Series(x) for x in node_list]
                VL.append(sum(series) / len(node_list))
        VLs.append(VL)
        
    print(f"Created communities' VLs")
    
    # Predictions for each iteration's VLs
    all_preds = []
    # It starts from 1 because we don't predict starting from the initial VLs of the ndoes
    for i in range(1, len(obj)):
        preds = {}
        for node in sorted(G.nodes()):
            try:
                # Get node community
                node_comm = obj[0][1][node - 1].index(1)
                # Get node VL
                node_VL = obj[i][1][node - 1]
                comms_VLs = VLs[i].copy()
                # Delete from communities' VLs the one related to the actual community of the node 
                del comms_VLs[node_comm]
                # Calculate the cosine similarities between the node VL and all the communities' VLs
                dists = [np.dot(node_VL, x)/(np.linalg.norm(node_VL)*np.linalg.norm(x)) for x in comms_VLs]
                if dists.index(max(dists)) >= node_comm:
                    # If the index of the maximum distance is greater or equal than the community of the node,
                    # the new one is defined by index + 1 because we have removed from comms_VLs
                    # the one related to the node.
                    # For example if we have removed community 5 and the index is 5, it needs to be incremented by 1.
                    preds[node] = dists.index(max(dists)) + 1
                else:
                    preds[node] = dists.index(max(dists))
            except (IndexError, ValueError):
                # Cases when we don't have a community for the node
                node_VL = obj[i][1][node - 1]
                comms_VLs = VLs[i].copy()
                dists = [np.dot(node_VL, x)/(np.linalg.norm(node_VL)*np.linalg.norm(x)) for x in comms_VLs]
                preds[node] = dists.index(max(dists))
        all_preds.append(preds)
    
    print("Completed predictions")
    
    # Find the candidates which changed their party
    data = pd.read_csv(f"./id_numerici/{folder}/nodelabels.csv")
    data2 = pd.read_csv(f"./id_numerici/{next_folder}/nodelabels.csv")
    
    changed = {}
    for node in G.nodes():
        # Party of the candidate in month x
        row = data[data["NodeID"] == node]
        # Party of the candidate in month x + 1
        row2 = data2[data2["NodeID"] == node]
        try:
            new_comm = row2["AssocLabelID"].values
            old_comm = row["AssocLabelID"].values
            if len(old_comm) != 0 and len(new_comm) != 0 and new_comm != (old_comm):
                # If there was a change of community memorize the old and new community
                # The -1 is put beacuse in the predictions we have the index of the community in [0, total_comms]
                changed[node] = (old_comm[0] - 1, new_comm[0] - 1)
        except ValueError: continue
            
    accuracies = []
    recall = []
    
    # Vec is a dictionary of predictions {node: new party}
    for vec in all_preds:
        rs = recall_score([x[1] for x in changed.values()], [vec[i] for i in changed.keys()], average="micro")
        recall.append(rs)
        ok = 0
        err = 0
        for node in changed.keys():
            if vec[node] == changed[node][1]: ok += 1
            else: err += 1
        try:        
            accuracies.append(ok / (ok + err))
        except ZeroDivisionError:
            accuracies.append(-1)
    
    counts.append(len(changed))
    all_accuracies.append(accuracies)
    iterations.append(accuracies.index(max(accuracies)))
    all_recalls.append(recall)
    
    print(counts)
    print(accuracies)
    
    print(f"Completed iteration in {time.time() - s_time}")
    total_preds.append(all_preds)
```

### Using second method

```python
total_preds2 = []
all_accuracies2 = []
all_preds2 = []
all_recalls2 = []
iterations2 = []
counts2 = []
for i in range(len(os.listdir("./id_numerici/")) - 1):
    
    # Get the two months to compare's folders
    folder = sorted(os.listdir("./id_numerici/"))[i]
    next_folder = sorted(os.listdir("./id_numerici/"))[i + 1]
    
    s_time = time.time()
    
    # Reading VLs from month folder
    obj = pd.read_pickle(f"./id_numerici/{folder}/results/log_trial_0_LPStates.pickled")
    
    # Reading initial parties of nodes exluding nodes without links
    comms = []
    for node in sorted(list(G.nodes)):
        if(1 in obj[0][1][node - 1]):
            comms.append(obj[0][1][node - 1].index(1))
        else: comms.append(-1)
    data = comms
    
    # Predictions for each iteration's VLs
    all_preds2 = []
    for i in range(1, len(obj)):
        preds = {}
        for node in sorted(G.nodes()):
            try:
                node_comm = obj[0][1][node - 1].index(1)
                node_VL = obj[i][1][node - 1].copy()
                del node_VL[node_comm]
                if node_VL.index(max(node_VL)) >= node_comm:
                    preds[node] = node_VL.index(max(node_VL)) + 1 
                else:
                    preds[node] = node_VL.index(max(node_VL))
            except ValueError:
                node_VL = obj[i][1][node - 1].copy()
                preds[node] = node_VL.index(max(node_VL))
        all_preds2.append(preds)
    
    print("Completed predictions")
    
    ### Checking changing values
    data = pd.read_csv(f"./id_numerici/{folder}/nodelabels.csv")
    data2 = pd.read_csv(f"./id_numerici/{next_folder}/nodelabels.csv")
    
    changed = {}
    for node in G.nodes():
        row = data[data["NodeID"] == node]
        row2 = data2[data2["NodeID"] == node]
        try:
            new_comm = row2["AssocLabelID"].values
            old_comm = row["AssocLabelID"].values
            if len(old_comm) != 0 and len(new_comm) != 0 and new_comm != (old_comm):
                changed[node] = (old_comm[0] - 1, new_comm[0] - 1)
        except ValueError: continue
            
    accuracies = []
    recall = []
    for vec in all_preds2:
        rs = recall_score([x[1] for x in changed.values()], [vec[i] for i in changed.keys()], average="micro")
        recall.append(rs)
        ok = 0
        err = 0
        for node in changed.keys():
            if vec[node] == changed[node][1]: ok += 1
            else: err += 1
        try:        
            accuracies.append(ok / (ok + err))
        except ZeroDivisionError:
            accuracies.append(-1)
    
    counts2.append(len(changed))
    all_accuracies2.append(accuracies)
    all_recalls2.append(recall)
    iterations2.append(accuracies.index(max(accuracies)))
    
    print(counts2)
    print(accuracies)
    
    print(f"Completed iteration in {time.time() - s_time}")
    total_preds2.append(all_preds2)
```

### Equal predictions graph

```python
mean_diff = []
it = len(total_preds[0])
months_num = len(total_preds)
for i in range(it):
    sum_eq = 0
    for j in range(months_num):
        t1 = list(total_preds[j][i].values())
        t2 = list(total_preds2[j][i].values())
        for k in range(len(t1)):
            if t1[k] == t2[k]:
                sum_eq += 1
    mean_diff.append(sum_eq / months_num)
```
